```python
def read_dataframe(filename):
    df = pd.read_parquet(filename)
    
    df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df.duration = df.duration.apply(lambda td: td.total_seconds() /60)
    
    df = df[(df.duration >=1) & (df.duration <=60)]
    
    caterogical =['PULocationID','DOLocationID']
    df[caterogical] = df[caterogical].astype(str)
    
    return df
    
```

```python
df_train =read_dataframe('/home/mpaprocki/mlops-zoomcamp1.2.2-1/notebooks/data/yellow_tripdata_2023-01.parquet')
df_val = read_dataframe('/home/mpaprocki/mlops-zoomcamp1.2.2-1/notebooks/data/yellow_tripdata_2023-02.parquet')
```

```python
len(df_train), len(df_val)
```

```python
caterogical =['PULocationID','DOLocationID']
numerical = ['trip_distance']

dv = DictVectorizer()

train_dicts = df_train[caterogical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dicts)

val_dicts = df_val[caterogical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dicts)
```

```python
target ='duration'
y_train = df_train[target].values
y_val = df_val[target].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```

```python
lr = Lasso()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```

```python
lr = Lasso(alpha=0.001)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```

```python
lr = Ridge(alpha=0.001)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_val)
mean_squared_error(y_val, y_pred, squared=False)
```
